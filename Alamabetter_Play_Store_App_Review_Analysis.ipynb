## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
#Convert install coloum string to int in install price and review

def convert_int(x) :
  if isinstance(x, int):
    return x
  x = str(x)
  x = x.replace(',','')
  x = x.replace('+', '')
  x = x.replace('M', '')
  x = x.replace('k', '')
  x = x.replace('$', '')
  if x == 'Free' or x == 'Varies with device' or x == 'Everyone':
    return 0
  else:
    return int(float(x))
````

````python
#Converting installs,price,reviews to int value for better calculations
df1['Installs']=df1['Installs'].apply(convert_int)
df1['Size_int']=df1['Size'].apply(convert_int)
df1['Price']=df1['Price'].apply(convert_int)
df1['Reviews']=df1['Reviews'].apply(convert_int)
````

````python
#convert size due to size divided into multiple category so we convert into to same format(mb=1024k, k=mb/1024)
def convert_int2(x):
    x=x.lower()

    if x=='varies with device':
        x=0
    else:
        if 'm' in x:
            x=x.replace('m','')
            x=eval(x)
        elif 'k' in x :
            x=x.replace('k','')
            x=eval(x)/1024
        else:
            x=-100

    return x
df1['Size_int']=df1['Size'].apply(convert_int2)
````

````python
# adjusting those app ratings which are  more than 5.
#rating_more_5 = df1[df1['Rating'] >5].index
rating_median = df1['Rating'].median()
df1.loc[df1['Rating']>5,'Rating'] = rating_median
````

````python

df1['Rating']=df1['Rating'].fillna(rating_median)
df1.isna().sum()
````

````python
#Removing the app which contains no type

type_null_index=df1[df1['Type'].isna()].index
df1=df1.drop(index=type_null_index)
df1.isnull().sum()
````

````python
df1 = df1.drop_duplicates(subset=['App'])
df1.shape
````
